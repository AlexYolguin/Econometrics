```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# const variables
num = 251 # number of observations
seed = 23 # random seed (if needs)
```

```python
def plot_data(
    data: np.array
    , title: str=None
    , style: str='b-'
) -> None:
    fig, ax = plt.subplots(figsize=(10, 5), constrained_layout=True)

    ax.set(title=title)
    ax.plot(data, style)

    plt.show()
```

```python
# Генерируем WN из стандартного N(0, 1)
#np.random.seed(seed)
wn = np.random.standard_normal(num)
print(f"WN: {wn[:10]}")

plot_data(wn, title="WN")
```

```python
# Random Walk
# Y_t = Y_(t-1) + Eps_t

# X_t - движение процесса в t момент времени (при X_t = N(0, 1))
#np.random.seed(seed)

X_t = np.random.normal(0, 1, size=num)
#X_t = [random.choice([-1, 1]) for _ in range(num)] # simple RW

print(f"X_t: {X_t[:10]}", end="\n\n")

rw = np.cumsum(X_t)
print(f"RW: {rw[:10]}")

plot_data(rw, title="RW")
```

```python
# Random walk with constant
#np.random.seed(seed)
def add_const(data: np.array, const: int | float) -> np.array:
    return data + np.array(const)

rw_with_const = add_const(rw, -1)
print(f"RW with const: {rw_with_const[:10]}")

plot_data(
    rw_with_const
    , title="Random Walk with consant"
    , style="g-"
)
```

```python
# Random Walk with trend
def add_trend(data: np.array, a: int | float = 1) -> np.array:
    num = len(data)
    
    # Y_t = Y_t + a * t 
    return data + (a * np.arange(num))

rw_with_trend = add_trend(rw, a=0.15)
print(f"RW with trend: {rw_with_trend[:10]}")

plot_data(
    rw_with_trend
    , title="rw_with_trend"
    , style="r-"
    )
```
